# Getting data

```python
import pandas as pd
import requests
import time
from datetime import datetime, timedelta
from stable_baselines3 import DQN
import gymnasium as gym
import gym_trading_env
import numpy as np
import plotly.graph_objs as go
import plotly.io as pio
import time
from gym_trading_env.downloader import download
from sklearn.preprocessing import StandardScaler
import os
```

```python
# List of cryptocurrency symbols
symbols = [
    "tBTCUSD",  # Bitcoin
    "tETHUSD",  # Ethereum
    "tLTCUSD",  # Litecoin
    "tXRPUSD",  # Ripple
    "tEOSUSD",  # EOS
    "tXMRUSD",  # Monero
    "tZECUSD",  # Zcash
    "tNEOUSD",  # Neo
]
timeframe = "1h"  # Hourly timeframe
limit = 10000  # Maximum limit

def fetch_data(symbol, start_date, end_date):
    url = f"https://api-pub.bitfinex.com/v2/candles/trade:{timeframe}:{symbol}/hist"
    print(url)
    params = {
        "limit": limit,
        "start": int(start_date.timestamp() * 1000),
        "end": int(end_date.timestamp() * 1000),
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        historical_candles = response.json()
        for candle in historical_candles:
            timestamp = candle[0] / 1000  # Convert milliseconds to seconds
            candle[0] = datetime.fromtimestamp(timestamp)
        df = pd.DataFrame(historical_candles, columns=['Timestamp', 'Open', 'Close', 'High', 'Low', 'Volume'])
        df.set_index('Timestamp', inplace=True)
        df.rename(columns={
            'Open': "open", 
            'Close': "close", 
            'High': "high", 
            'Low': "low", 
            'Volume': "Volume USD"
        }, inplace=True)
        df = df.sort_index(ascending=True)
        return df
    else:
        print(f"Failed to fetch data for {symbol}: {response.status_code} - {response.reason}")
        return pd.DataFrame()  # Return an empty DataFrame in case of error

date_ranges = [
    (datetime(2016, 9, 8), datetime(2017, 9, 8)),
    (datetime(2017, 9, 8), datetime(2018, 9, 8)),
    (datetime(2018, 9, 8), datetime(2019, 9, 8)),
    (datetime(2019, 9, 8), datetime(2020, 9, 8)),
    (datetime(2020, 9, 8), datetime(2021, 9, 8)),
    (datetime(2021, 9, 8), datetime(2022, 9, 8)),
    (datetime(2022, 9, 8), datetime(2023, 9, 8)),
    (datetime(2023, 9, 8), datetime(2024, 9, 8)),
]

# Loop through each symbol
for symbol in symbols:
    # Initialize empty DataFrame to concatenate data for each symbol
    all_data = pd.DataFrame()

    # Loop through the date ranges and fetch data for the current symbol
    for start_date, end_date in date_ranges:
        df = fetch_data(symbol, start_date, end_date)
        all_data = pd.concat([all_data, df])
        time.sleep(5)

    # Remove duplicates based on the index (Timestamp)
    all_data = all_data[~all_data.index.duplicated(keep='first')]

    # Reset index to ensure proper ordering
    all_data = all_data.sort_index(ascending=True)

    # Save data to a CSV file under the 'data' directory
    filename = f"data/{symbol.replace('t', '')}.csv"  # Save with a simpler name
    all_data.to_csv(filename)
    print(f"Data for {symbol} saved to {filename}")
```

# Adding indicators

```python
def add_sma(df, period=14):
    df = df.copy() 
    df.loc[:, f'feature_sma_{period}'] = df['close'].rolling(window=period).mean()
    return df

def add_ema(df, period=14):
    df = df.copy() 
    df.loc[:, f'feature_ema_{period}'] = df['close'].ewm(span=period, adjust=False).mean()
    return df

def add_rsi(df, period=14):
    df = df.copy()  
    delta = df['close'].diff(1)
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    avg_gain = gain.rolling(window=period).mean()
    avg_loss = loss.rolling(window=period).mean()
    rs = avg_gain / avg_loss
    df.loc[:, f'feature_rsi_{period}'] = 100 - (100 / (1 + rs))
    return df

def add_macd(df, short_period=12, long_period=26, signal_period=9):
    df = df.copy()  
    short_ema = df['close'].ewm(span=short_period, adjust=False).mean()
    long_ema = df['close'].ewm(span=long_period, adjust=False).mean()
    df.loc[:, 'feature_macd'] = short_ema - long_ema
    df.loc[:, 'feature_macd_signal'] = df['feature_macd'].ewm(span=signal_period, adjust=False).mean()
    return df
def add_bollinger_bands(df, period=20, num_std_dev=2):
    sma = df['close'].rolling(window=period).mean()
    std_dev = df['close'].rolling(window=period).std()
    df.loc[:, 'feature_bollinger_upper'] = sma + (std_dev * num_std_dev)
    df.loc[:, 'feature_bollinger_lower'] = sma - (std_dev * num_std_dev)
    return df
def add_obv(df):
    df.loc[:, 'feature_obv'] = (np.sign(df['close'].diff()) * df['Volume USD']).fillna(0).cumsum()
    return df
```

# Train/Test split and train the DQN model

```python
# Initialize the scaler
scaler = StandardScaler()

# Function to process the data for each cryptocurrency
def process_crypto_data(file_path):
    # Load the data
    all_data = pd.read_csv(file_path, index_col='Timestamp', parse_dates=True)
    
    # Add indicators
    all_data = add_sma(all_data, period=14)
    all_data = add_ema(all_data, period=14)
    all_data = add_rsi(all_data, period=14)
    all_data = add_macd(all_data)
    all_data = add_bollinger_bands(all_data)
    all_data = add_obv(all_data)
    all_data = all_data.dropna()
    
    # Select the feature columns to scale
    feature_columns = [col for col in all_data.columns if col.startswith('feature_')]
    
    # Fit the scaler on the feature columns and transform
    all_data[feature_columns] = scaler.fit_transform(all_data[feature_columns])
    
    percent_split = 0.2
    split_index = int(len(all_data) * (1 - percent_split))
    df_training = all_data.iloc[:split_index].copy()
    df_testing = all_data.iloc[split_index:].copy()

    return df_training, df_testing
```

```python
# List to hold training and testing data for each cryptocurrency
train_test_data = []

   

# List of CSV files for each cryptocurrency
crypto_files = ["BTCUSD.csv", "ETHUSD.csv", "LTCUSD.csv", "EOSUSD.csv", "NEOUSD.csv", "XMRUSD.csv", "ZECUSD.csv"]

# Directory where the data is stored
data_directory = "data"

# Loop through the cryptocurrency files and process each one
for file_name in crypto_files:
    file_path = os.path.join(data_directory, file_name)
    df_training, df_testing = process_crypto_data(file_path)
    train_test_data.append((df_training, df_testing))
```

```python
# Define directories for saving the processed data
train_dir = 'data/train'
test_dir = 'data/test'

# Create directories if they do not exist
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Loop through the train_test_data list and save DataFrames
crypto_names = ["BTCUSD", "ETHUSD", "LTCUSD", "EOSUSD", "NEOUSD", "XMRUSD", "ZECUSD"]
for (df_training, df_testing), crypto in zip(train_test_data, crypto_names):
    # Save training data
    training_file_path = os.path.join(train_dir, f"{crypto}_training.pkl")
    df_training.to_pickle(training_file_path)
    
    # Save testing data
    testing_file_path = os.path.join(test_dir, f"{crypto}_testing.pkl")
    df_testing.to_pickle(testing_file_path)
```

```python
import matplotlib.pyplot as plt
from collections import Counter

done, truncated = False, False
observation, info = env.reset()
print(info)
portfolio_values = []  # List to store portfolio values at each step
rewards = []           # List to store rewards
actions = []           # List to store actions

while not done and not truncated:
    action, _states = model.predict(observation)  # Predict the action
    observation, reward, done, truncated, info = env.step(action)  # Take a step in the environment

    portfolio_values.append(info["portfolio_valuation"])  # Assuming 'portfolio_value' is in info
    rewards.append(reward)
    actions.append(action.item())

# Plot portfolio values
plt.figure(figsize=(10, 5))
plt.plot(portfolio_values, label="Portfolio Value", color='b')
plt.title("Portfolio Value Over Time")
plt.xlabel("Step")
plt.ylabel("Portfolio Value")
plt.legend()
plt.show()

# Plot rewards
plt.figure(figsize=(10, 5))
plt.plot(rewards, label="Rewards", color='g')
plt.title("Rewards Over Time")
plt.xlabel("Step")
plt.ylabel("Reward")
plt.legend()
plt.show()

# Plot actions
plt.figure(figsize=(20, 10))
plt.plot(actions, label="Actions", color='r')
plt.title("Actions Taken Over Time")
plt.xlabel("Step")
plt.ylabel("Action")
plt.legend()
plt.show()
action_counts = Counter(actions)
print("Action counts:", action_counts)
```
